The following code is used to scrape the data of MBA colleges in Bengaluru from shiksha.com

2) Importing required pacakages 

```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
```

3) Initialising the required variables required

```python
n=10
count=0
College_name=[]
location=[]
ratings=[]
# other_details=[]
headers={
    "Accept-Language" : "en-US,en;q=0.9",
    "upgrade-insecure-requests":"1",
    "dnt":"1",
    "User-Agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/111.0.0.0 Safari/537.36 Edg/111.0.1661.41",
    "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,image/webp,image/apng,*/*;q=0.8,application/signed-exchange;v=b3;q=0.7"
}
```

4) Loop for sending request and collecting the required data

    We use a loop here because a single request only loads a small amount of data, and we need to send multiple requests to get all the data.

```python
for i in range(n):
    
    if i == 0:
        url = "https://www.shiksha.com/mba/colleges/mba-colleges-bangalore"
    else:
        url = "https://www.shiksha.com/mba/colleges/mba-colleges-bangalore-" + str(i)
    r=requests.get(url,headers=headers)

    print(url)
    # print(r.text[:500])

    soup=BeautifulSoup(r.text,"html.parser")
    # print(soup.prettify())

    for section in soup.findAll('div', attrs={"class":'ctpSrp-contnr'}):
        college= section.findAll('div',attrs={'class':"_1822 _0fc7"})

        
        for c in college:
            
            name=c.find('div', attrs={"class":"c8ff"})
            print(name.text)
            name = name.text

            loc=c.find('span', attrs={"class":"_5588"})
            print(loc.text)
            loc = loc.text
            
        
            rating=c.find('a', attrs={"class":"_68c4 ripple dark"})
            if(hasattr(rating,'text')):
                print(rating.text)
                rating = rating.text
            else:
                rating = "No rating"
                print("No Rating")
            
            # d=c.find('div', attrs={"class":"ctp-detail"})
            # print(d.text)

            if (name in College_name) and (loc in location):
                break
            else:
                count= count+1
                College_name.append(name)
                location.append(loc)
                ratings.append(rating)
                # other_details.append(text)
                
            print()
    
```

5) To verify that all lists have the same number of data points and to count how many data points we have collected.

```python
print(count)
print(len(College_name))
print(len(location))
print(len(ratings))
# print(len(other_details))
```

6) Creating DataFrame and save it as a CSV file and print the data

```python
dict = {'Institution_Name': College_name, 'Location': location, 'Ratings': ratings}
data = pd.DataFrame(dict)

print(data.head())
data.to_csv('list_of_MBA_colleges_bangalore.csv', sep='\t', index=True,header=True)
```
